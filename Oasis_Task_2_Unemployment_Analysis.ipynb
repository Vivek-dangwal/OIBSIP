```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Load the dataset (Oasis usually provides 'Unemployment_in_India.csv')
df = pd.read_csv('Unemployment in India.csv')

# Display first few rows to confirm it's loaded
df.head()
```

```python
# 1. Remove extra spaces from column names
df.columns = df.columns.str.strip()

# 2. Convert 'Date' to a format Python understands
df['Date'] = pd.to_datetime(df['Date'])

# 3. Add a Month column (this helps for better analysis later)
df['Month'] = df['Date'].dt.month_name()

# 4. Drop any empty rows that might be at the bottom of the CSV
df.dropna(inplace=True)

print("Data cleaning is done! Here are the clean columns:")
print(df.columns)
```

```python
import plotly.express as px

# Create an interactive bar chart
fig = px.bar(df, x='Region', y='Estimated Unemployment Rate (%)',
             color='Region', title='Unemployment Rate by Region',
             template='plotly_dark')

fig.show()
```

```python
# Create an interactive line chart to see the trend over time
fig = px.line(df, x='Date', y='Estimated Unemployment Rate (%)',
              color='Region', title='Unemployment Rate Trend in India (2020)',
              template='plotly_dark')

# Add a vertical line to mark the start of the lockdown (March 2020)
fig.add_vline(x='2020-03-24', line_width=3, line_dash="dash", line_color="red")
fig.add_annotation(x='2020-03-24', text="Lockdown Starts", showarrow=True, arrowhead=1)

fig.show()
```

```python
# Box plot to compare Rural vs Urban unemployment
fig = px.box(df, x='Area', y='Estimated Unemployment Rate (%)',
             color='Area', title='Urban vs. Rural Unemployment Rate',
             points="all", template='plotly_white')

fig.show()
```
